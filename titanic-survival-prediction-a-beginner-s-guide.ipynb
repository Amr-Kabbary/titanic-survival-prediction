### Handling Missing Values

In this section, we check for missing values in the cleaned dataset using the `isna().sum()` method. This helps us identify which columns have missing data and how much, so we can decide how to handle them (e.g., imputation or removal).

```python
# Check for missing values in the cleaned dataset
train_data.isna().sum()
```

```python
train_data.dropna(subset=['Embarked'], inplace=True)

# Impute missing 'Age' with median
train_data['Age'].fillna(train_data['Age'].median(), inplace=True)
test_data['Age'].fillna(test_data['Age'].median(), inplace=True)

train_data.isna().sum()
```

```python
train_data[['cabin_char', 'cabin_num']] = train_data['Cabin'].str.extract(r'([a-zA-Z]+)(\d+)')
test_data[['cabin_char', 'cabin_num']] = test_data['Cabin'].str.extract(r'([a-zA-Z]+)(\d+)')

train_data.head(10)
```

```python
train_data['Title'] = train_data['Name'].str.extract(' ([A-Za-z]+)\.', expand=False)
train_data = train_data.drop(['Name','Ticket'],axis=1)

test_data['Title'] = test_data['Name'].str.extract(' ([A-Za-z]+)\.', expand=False)
test_data = test_data.drop(['Name','Ticket'],axis=1)

train_data.head(10)
```

```python
train_data["Fam_size"] = train_data["SibSp"] + train_data["Parch"] + 1 
test_data["Fam_size"] = test_data["SibSp"] + test_data["Parch"] + 1 

train_data['AgeGroup'] = pd.cut(train_data['Age'], bins=[0, 18, 35, 50, 100], labels=['Child', 'Young Adult', 'Adult', 'Senior'])
test_data['AgeGroup'] = pd.cut(test_data['Age'], bins=[0, 18, 35, 50, 100], labels=['Child', 'Young Adult', 'Adult', 'Senior'])

train_data['AgeGroup'] = train_data['AgeGroup'].astype(str)
test_data['AgeGroup'] = test_data['AgeGroup'].astype(str)

test_data.head(10)
```

### Preparing Data for Modeling

In this section, we prepare the data for machine learning by:
1. Splitting the cleaned dataset into training and validation sets.
2. Identifying numerical and categorical columns.
3. Ensuring consistency across training, validation, and test datasets.

```python
# Import the necessary library
from sklearn.model_selection import train_test_split

# Define the target variable (y) and features (X)
y = train_data.Survived
X = train_data.drop('Survived', axis=1)

# Split the data into training and validation sets
X_train_full, X_valid_full, y_train, y_valid = train_test_split(X, y, train_size=0.8, test_size=0.2, random_state=0)


# Combine numerical and categorical columns
num_col = [cname for cname in X.columns 
           if train_data[cname].dtype in ['int64','float64']]
cat_col = [cname for cname in X.columns 
           if train_data[cname].dtype == 'object']

print(num_col)
print(cat_col)

my_col = num_col + cat_col

# Ensure consistency across training, validation, and test datasets
X_train = X_train_full[my_col].copy()
X_valid = X_valid_full[my_col].copy()
X_test = test_data[my_col].copy()
```

```python
X_valid.head()
```

### Data Preprocessing Pipeline

In this section, we create a preprocessing pipeline to handle missing values and encode categorical variables. The pipeline consists of:
1. **Numerical Columns**: Imputing missing values with a constant strategy.
2. **Categorical Columns**: Imputing missing values with the most frequent value and one-hot encoding the categories.

```python
# Import necessary libraries
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder

# Define the numerical transformer
num_transformer = SimpleImputer(strategy='median')

# Define the categorical transformer as a pipeline
cat_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),  # Impute missing values
    ('onehot', OneHotEncoder(handle_unknown='ignore'))    # One-hot encode categories
])

# Combine numerical and categorical transformers into a ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
        ('num', num_transformer, num_col),  # Apply to numerical columns
        ('cat', cat_transformer, cat_col)   # Apply to categorical columns
    ])
```

### Model Definition

In this section, we define a Random Forest Classifier model. Random Forest is an ensemble learning method that combines multiple decision trees to improve predictive performance and reduce overfitting. We use 600 estimators and set a random state for reproducibility.

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(random_state=42,C=1,solver='lbfgs')
```

### Model Training and Evaluation

In this section, we:
1. Combine the preprocessing steps and the Random Forest model into a single pipeline.
2. Train the pipeline on the training data.
3. Make predictions on the validation set.
4. Evaluate the model's performance using accuracy scores for both the training and validation sets.

```python
# Import the accuracy_score metric
from sklearn.metrics import accuracy_score

# Create a pipeline combining preprocessing and the model
my_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),  # Apply preprocessing
    ('model', model)                # Apply the Random Forest model
])

# Train the pipeline on the training data
my_pipeline.fit(X_train, y_train)

# Make predictions on the validation set
preds = my_pipeline.predict(X_valid)

# Evaluate the model
print('The training score for LogisticRegression is:', (my_pipeline.score(X_train, y_train) * 100), '%')
print('Validation accuracy:', (accuracy_score(y_valid, preds) * 100), '%')
```

### Generating Predictions and Creating Submission File

In this section, we:
1. Use the trained pipeline to make predictions on the test dataset.
2. Create a submission file in the required format (with `PassengerId` and `Survived` columns).
3. Save the submission file as a CSV for uploading to Kaggle.

```python
# Generate predictions for the test dataset
preds_final = my_pipeline.predict(X_test)

# Create a DataFrame for the submission file
output = pd.DataFrame({'PassengerId': test_data.PassengerId, 'Survived': preds_final})

# Save the submission file as a CSV
output.to_csv('submission.csv', index=False)
print("Your submission was successfully saved!")
```
